## Kaks results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## read KaKs results
kaks = "tong_bhlh_pan_gene_pairs_updated.kaks"
df_kaks = pd.read_csv(kaks,header=0,sep='\t')
df_kaks2 = df_kaks[df_kaks["Ka/Ks"] <10] ## filter kaks outlier to get more sensible results
df_kaks2.head()
```

```python
##split gene pair IDs into two columns: gene1, gene2
df_kaks2[['gene1', 'gene2']] = df_kaks2['Sequence'].str.split('-', expand=True)
```

```python
df_kaks2.head()
```

```python
## use groupby to calcumate mean KaKs for each bHLH gene OG
df_kaks_mean = df_kaks2.groupby("gene1").mean()
df_kaks_mean.head()
```

```python
df_match = pd.read_csv("tongcen_bhlh_v2_id_list_match_panMorex.txt",delim_whitespace=True,header=None)
#df_match = df_match.astype("str")
df_match = df_match.replace("\.1$","",regex=True) ## remove transcript id
df_match.columns = ["V1","V2"]
df_match.head()
```

```python
## only use those bHLH genes with domseq >30aa
df_seq30aa = pd.read_csv("domseq.id.txt",header=None)
df_seq30aa.columns = ["seqV2"]
df_seq30aa.head()
```

```python
df_seq30aa.shape
```

```python
## match V2 IDs with pan Morex IDs
df_merged2 = df_match.merge(df_seq30aa,left_on="V2",right_on="seqV2",how="left")
df_merged2 = df_merged2.merge(df_kaks_mean,left_on="V1",right_index=True,how="left")
df_merged2.head()
```

```python
cols = ["V1","V2","Ka","Ks","Ka/Ks","P-Value(Fisher)","Length","Substitutions","S-Substitutions","N-Substitutions"]
```

```python
## remove empty rows and sort out column order
df_merged2 = df_merged2[df_merged2["seqV2"].notna()][cols]
```

```python
## add transcript id back
df_merged2["V2"] = df_merged2["V2"].astype("str") + ".1"
```

```python
df_merged2.head()
```

```python
## read GRR results
df_GRR = pd.read_csv("MOREX_gene_GRR.txt",header=None,sep=" ")
df_GRR.columns = ["V2","GRR"]
df_GRR.head()
```

```python
## GRR results gene IDs derived from CD-HITs parsing, need to be modifed to match with orginal gene IDs in KaKs data
df_GRR["V2"] = df_GRR["V2"].str.replace("MOREXr2","MOREX_r2")
df_GRR["V2"] = df_GRR["V2"].str.replace("_",".")
df_GRR.head()
```

```python
## merge GRR and KaKs data
df_kaks_GRR = df_merged2.merge(df_GRR,on="V2",how="left")
df_kaks_GRR.head()
```

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    numbin = 10
    binwidth = xymax/numbin
    #lim = (int(xymax/binwidth) + 1) * binwidth
    lim = xymax + binwidth

    bins = np.arange(0, lim, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(df.Ka, df.Ks, ax, ax_histx, ax_histy)

plt.show()
```
